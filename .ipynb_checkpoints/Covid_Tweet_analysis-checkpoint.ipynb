## Pre-Defined Functions

```python
def fetchweek(date):
    try:
        date =datetime.strptime(date,'%d-%m-%Y')
        if date < datetime(2020,3,22):
            return 'Week 12'
        elif date >= datetime(2020,3,23) and date <= datetime(2020,3,29):
            return 'week_13'
        elif date >= datetime(2020,3,30) and date <= datetime(2020,4,5):
            return 'week_14'
        elif date >= datetime(2020,4,6) and date <= datetime(2020,4,12):
            return 'week_15'
        elif date >= datetime(2020,4,13) and date <= datetime(2020,4,19):
            return 'week_16'    
        elif date >= datetime(2020,4,20) and date <= datetime(2020,4,26):
            return 'week_17'
        elif date >= datetime(2020,4,27) and date <= datetime(2020,5,3):
            return 'week_18'
        elif date >= datetime(2020,5,4) and date <= datetime(2020,5,10):
            return 'week_19'
        elif date >= datetime(2020,5,11) and date <= datetime(2020,5,17):
            return 'week_20'       
        elif date >= datetime(2020,5,18) and date <= datetime(2020,5,24):
            return 'week_21'   
        elif date >= datetime(2020,5,25) and date <= datetime(2020,5,31):
            return 'week_22' 
        else:
            return 'week_25'
    except:
        return 'week_25'

def categorizeLexiconWords(df):
    cat = ["positive_emotion","negative_emotion","hate","love","fun","joy","sadness","anger","envy","horror","pain","shame","nervousness","disgust","fear","irritate"]
    d = {el:0 for el in cat}
    for index,row in df.iterrows():
        try:
            derdict = Empath().analyze(row['tweet'], categories=cat)
            df.loc[index,'emotion'] = [k for k in set(derdict) if derdict.get(k) > 0][0]
        except:
            df.loc[index,'emotion'] = "neutral"
#         df.loc[index,'positive_emotion'] = derdict['positive_emotion']
#         df.loc[index,'negative_emotion'] = derdict['negative_emotion']
#         df.loc[index,'hate'] = derdict['hate']
#         df.loc[index,'love'] = derdict['love']
#         df.loc[index,'fun'] = derdict['fun']
#         df.loc[index,'joy'] = derdict['joy']
#         df.loc[index,'sadness'] = derdict['sadness']
#         df.loc[index,'anger'] = derdict['anger']
#         df.loc[index,'envy'] = derdict['envy']
#         df.loc[index,'horror'] = derdict['horror']
#         df.loc[index,'violence'] = derdict['violence']
#         df.loc[index,'pain'] = derdict['pain']
#         df.loc[index,'shame'] = derdict['shame']
#         df.loc[index,'nervousness'] = derdict['nervousness']
#         df.loc[index,'disgust'] = derdict['disgust']
#         df.loc[index,'fear'] = derdict['fear']
#         df.loc[index,'irritate'] = derdict['irritate']  
    return df

def clean_tweets(tweet):
    stop_words = set(STOPWORDS)
    alphabets = set(string.ascii_lowercase)
    tweet = tweet.lower()
    #after tweet preprocessing the colon symbol left remain
    tweet = re.sub(r'‚Ä¶', '', tweet)
    #replacing the user tagging 
    tweet = re.sub(r'@([^\s]+)', ' ', tweet)
    #replace the URLs
    tweet = re.sub(r'http[s]?[:]?[\s]?\/\/(?:[a-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+',' ',tweet)
    #replace the n't
    tweet = re.sub(r'n\'t', ' not', tweet)
    #replace the 's'
    tweet = re.sub(r'\'s', '', tweet)
    #replace the numbers
    tweet = re.sub(r'\d', '', tweet)
    #replace the dots
    tweet = re.sub(r'…', '', tweet)

    exclude = set(string.punctuation).union(set(['’','…','``','...','“','‘','”']))

    word_tokens = word_tokenize(tweet)
    #filter using NLTK library append it to a string
    filtered_tweet = [w for w in word_tokens]
    filtered_tweet = []
    #looping through conditions
    for w in word_tokens:
    #check tokens against stop words and punctuations
        if w not in exclude and w not in ('rt','…','➡️') and w not in stop_words and w not in alphabets:
            filtered_tweet.append(w)
    return ' '.join(filtered_tweet)

```

## Topic Modelling

## Topics per Week-wise for work tweets

```python
Work_tweets_df = pd.read_csv('Data/TweetMar20_May26onlywork.csv')
Work_tweets_df.drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],axis=1,inplace=True)
week_list = list(Work_tweets_df.week.value_counts().index)
for current_week in week_list[:]:
    print('\033[1m'+current_week+'\033[0m')
    Work_tweets_df_pw = Work_tweets_df[Work_tweets_df['week']==current_week]
    #print(Work_tweets_df_pw.shape)
    Work_tweets_df_pw.tweet = Work_tweets_df_pw.tweet.map(clean_tweets)
    texts = [word_tokenize(Work_tweets_df_pw['tweet'][i]) for i,row in Work_tweets_df_pw.iterrows()]
    frequency = defaultdict(int)
    for text in texts:
        for token in text:
            frequency[token] += 1
    texts = [[token for token in text if frequency[token] >= 20] for text in texts]
    #First created a mapping for words and some index value
    dictionary = corpora.Dictionary(texts)
    #Below we are taking the Frequency of the each word as per the word's assigned index value above
    corpus = [dictionary.doc2bow(text) for text in texts]
    # Training LDA Model
    lda_model = models.ldamodel.LdaModel(corpus=corpus,id2word=dictionary,
                                               num_topics=10, 
                                               random_state=42,
                                               update_every=1,
                                               chunksize=200,
                                               passes=40,
                                               alpha='auto',
                                               eval_every=1,
                                               per_word_topics=True)
    lda_topics = lda_model.show_topics(num_words=5)
    topics = []
    filters = [lambda x: x.lower(), strip_punctuation, strip_numeric]

    for topic in lda_topics:
        #print(topic)
        topics.append(preprocess_string(topic[1], filters))
        
    print(*topics,sep='\n')

```
