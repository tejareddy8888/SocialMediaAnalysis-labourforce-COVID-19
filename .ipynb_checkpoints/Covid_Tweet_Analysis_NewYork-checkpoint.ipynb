## Pre-function Declaration

```python
def load_data_fromfile(file):
    data = pd.read_csv(file,header=None,names=['tweet_id','sentiment'])
    return data

def fetch_file_fromdir(directory):
    df = pd.DataFrame({'tweet_id' : [],'sentiment': []})
    for file_path in tqdm(os.listdir(directory), desc=os.path.basename(directory)):
        df1 = load_data_fromfile(os.path.join(directory, file_path))
        df = pd.concat([df,df1]).reset_index(drop=True)
    return df

def fetch_and_load_datasets(directory):
    train_df = fetch_file_fromdir(os.path.join(os.getcwd(),directory))
    return train_df

def categorizeLexiconWords(df):
    """ args:
    input: DataFrame
    output: DataFrame
    Actions: uses the Empath library and read individual tweet and categorizes into multiple emotions as per emotion_category"""
    
    emotion_category = ["positive_emotion","negative_emotion","hate","love","fun","joy","sadness","anger","envy","horror","pain","shame","nervousness","disgust","fear","irritate"]
    
    emotion_dict = {single_emotion:0 for single_emotion in emotion_category}
    
    for index,row in df.iterrows():
        try:
            derdict = Empath().analyze(row['tweet'], categories=emotion_category)
            df.loc[index,'emotion'] = [k for k in set(derdict) if derdict.get(k) > 0][0]
        except:
            df.loc[index,'emotion'] = "neutral"
    return df

def sentimentClassification(df):
    for index,row in df.iterrows():
        try:
            if row['Sentiment'] >= 0:
                df.loc[index,'sent_class'] = 'pos'
            else:
                df.loc[index,'sent_class'] = 'neg'
        except:
            df.loc[index,'sent_class'] = 'na'
    return df
```

# Main work: Starts from here 

## Loading data

```python
data = pd.read_csv('Data/Tweets_Geotagged.csv',header=0,names=['Tweet_id','Sentiment'])
data.Tweet_id = data.Tweet_id.astype('int64')
Only_Newyork = pd.read_csv('Data/OnlyNewYorkTweets.csv')
Only_Newyork = Only_Newyork[['created_at','id','full_text','place']]
```

## merging for sentiment

```python
Only_Newyork = Only_Newyork.merge(data,how='inner',left_on='id',right_on='Tweet_id')
Only_Newyork.drop('id',axis=1,inplace=True)
```

## changing the timezone

```python
Only_Newyork.created_at = pd.to_datetime(Only_Newyork.created_at)
my_timezone = pytz.timezone('America/New_York')
Only_Newyork['created_at'] = Only_Newyork['created_at'].dt.tz_convert(my_timezone)
```

```python
Only_Newyork
```

## Categorize Emotions

```python
Only_Newyork = categorizeLexiconWords(Only_Newyork)
```

```python
#created weekday column
Only_Newyork['weekday']= Only_Newyork.created_at.dt.strftime('%A')
#created hour column
Only_Newyork['hour'] = Only_Newyork.created_at.dt.strftime('%H')
```

## Classify Sentiment into 3 Class

```python
Only_Newyork = sentimentClassification(Only_Newyork)
```

## Hourly mean sentiment for work-related Tweets New York Twitter Data 


```python
Only_Newyork.loc[:,['Sentiment','hour','weekday','sent_class']].groupby(['hour','sent_class']).mean().unstack(level=1,fill_value=0).rename(columns={"pos": "Positive", "neg": "Negative"})
```

```python
width=0.3
fig,ax = plt.subplots(sharex=True,figsize=(30,20))
Only_Newyork.loc[:,['Sentiment','hour','weekday','sent_class']].groupby(['hour','sent_class']).mean().unstack(level=1,fill_value=0).rename(columns={"pos": "Positive", "neg": "Negative"})[6:].plot.bar(ax=ax,width=width,stacked=True,color=['#FF0000','#008000'],label='Positive Sentiment')
ax.legend(bbox_to_anchor=(1.02,0), loc="lower left",prop={'size': 32})
ax.tick_params(labelrotation=45)
ax.xaxis.set_tick_params(labelsize=32)
ax.yaxis.set_tick_params(labelsize=32)
ax.set_title('Work Tweet Mean Sentiment Hour_Wise of US',fontdict={'fontsize':36})
ax.set_xlabel('Hour of the day', fontsize=32)
ax.set_ylabel('Mean Sentiment', fontsize=32)
```
